```python
from graphviz import Digraph
```

```python
dot = Digraph()
dot.attr(rankdir='LR')
# 普通状态
dot.attr('node',shape='circle')
dot.node('q0')
dot.node('q1')

# 终结状态
dot.attr('node',shape='doublecircle')
dot.node('q2')

dot.edge('q0','q0',label='0')
dot.edge('q0','q1',label='0')
dot.edge('q1','q2',label='1')
dot
```
